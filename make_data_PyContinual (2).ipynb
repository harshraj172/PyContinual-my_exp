```python
import pandas as pd
import json 
import os
import random
```

```python
data = pd.read_csv("/content/Dynamically Generated Hate Dataset v0.2.3 (1).csv")
data = data[data['target'].notna()]                                             
```

```python
data.head()
```

```python
def get_dict(data, type_, split, domain):
    type_data_hate = data[(data[domain].str.contains(type_)) & (data['split']==split)]
    type_data_nothate = data[(data[domain]=='none') & (data['split']==split)]
    type_data_nothate = type_data_nothate.sample(n=min(len(type_data_nothate), len(type_data_hate)))
    type_data = pd.concat([type_data_hate, type_data_nothate], axis=0).reset_index(drop=True)
    type_data = type_data.sample(frac = 1)

    dict_m = {}
    for i in range(len(type_data)):
        dict_ = {
            'id' : int(type_data['X1'][i]),
            'polarity' : "1" if type_data['label'][i]=='hate' else "-1",
            'sentence' : type_data['text'][i]
        }
        dict_m[i] = dict_
    print(dict_m)
    
    os.makedirs(f"./ssc/domain-{domain}/{type_}", exist_ok=True)
    with open(f"./ssc/domain-{domain}/{type_}/{split}.json", "w") as file:
        json.dump(dict_m, file)
```

```python
domain = 'target'

if domain == 'target':
    types = []
    for t in data['target'].dropna().unique():
        types.extend(t.split(', '))
    types = [t for t in set(types) if t!= 'notgiven' and t!= 'none']
else:
    types = [t for t in data['target'].unique() if t!= 'notgiven' and t!= 'none']
```

```python
for type_ in types:
    if type_ != 'notgiven' and type_ != 'none':
        print(type_)
        get_dict(data, type_=type_, split='train', domain=domain)
        get_dict(data, type_=type_, split='dev', domain=domain)
        get_dict(data, type_=type_, split='test', domain=domain)
```

```python
# to get a txt file with randomly shiffled task ids
with open(f'ssc_random-{domain}', 'w') as file:
    for _ in range(10):
        random.shuffle(types)
        file.writelines(
            '\t'.join(types) + '\n'
            )
```
